```python
import numpy as np
from numba import autojit
import matplotlib.pyplot as plt
```

```python
@autojit
def mandelbrot(real, imaginery, max_iter):
    c = complex(real, imaginery)
    z = 0.0j
    
    for i in range(max_iter):
        z = z*z + c
        if z.real*z.real + z.imag*z.imag >= 10:
            return 1
        
    return max_iter
```

```python
columns = 2000
rows = 2000

result = np.zeros([rows, columns])
for row_index, real in enumerate(np.linspace(-2, 1, num=rows)):
    for column_index, imaginery in enumerate(np.linspace(-1, 1, num=columns)):
        result[row_index, column_index] = mandelbrot(real, imaginery, 100)
            
plt.figure(dpi=100)
plt.imshow(result.T, cmap='hot', interpolation='bilinear', extent=[-2, 1, -1, 1])
plt.xlabel('Real')
plt.ylabel('Imaginery')
plt.title('Mandelbrot Set')
plt.show()
```
